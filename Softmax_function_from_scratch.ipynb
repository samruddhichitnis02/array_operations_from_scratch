### Formula for Softmax Function

- Here if there are 5 classes lets say, then
- e ^ zi is the exponential of the class you want the softmax value of 
- Sum(e ^ zj) is the sum of exponential value of all the 5 classes

### Points to Note about Softmax

- Softmax Function converts real values in to probabilities
- It is only used as output layer of neural network
- When there are only 2 classes, we can use Sigmoid activation
- When there are more than 2 classes at that time sigmoid activation functions are helpful

### Softmax implementation from scratch using Python

```python
import numpy as np
```

```python
x_input = np.array([-2, 4, 13, 25])
```

```python
Numerator = np.exp(x_input)
Numerator
```

```python
Denominator = np.sum(np.exp(x_input))
Denominator
```

```python
softmax_function = Numerator / Denominator
softmax_function
```

```python
np.argmax(softmax_function)
```

```python
def softmax(inp_array : np.array) -> int:
    Numerator = np.exp(inp_array)
    Denominator = np.sum(np.exp(x_input))
    softmax_output = Numerator / Denominator
    return np.argmax(softmax_output)
```

```python
x_input = np.array([-2, 4, 13, 25])
output_class = softmax(x_input)
print(output_class)
```

### Difference between Sigmoid and Softmax

- The difference between the two is that in the denominator of softmax, Softmax sums together all values, while Sigmoid just makes the output between 0 to 1.
- The sume of all Softmax units is supposed to be 1, but in Sigmoid, its not necessarily 1.
