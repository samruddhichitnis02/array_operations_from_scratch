### Importing required libraries

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) 
```

```python
x1 = np.linspace(-10, 10, 100) 
z = sigmoid(x1)

plt.plot(x1, z) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X)") 
plt.show()
```

```python
x2 = np.linspace(-100, 100, 200) 
z2 = sigmoid(x2)

plt.plot(x2, z2) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(X)") 
plt.show()
```
